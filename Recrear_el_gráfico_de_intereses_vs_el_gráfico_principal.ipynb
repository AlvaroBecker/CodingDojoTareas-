Creen un filtro para que el nombre de la hipoteca sea de “30 años”. Usen len() para encontrar cuántas filas corresponden a 30 años.

```python
thty_year=df['Mortgage Name']== '30 Year'
len(df.loc[(thty_year),['Mortgage Name']])
```

Creen un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).

```python
tres_porciento=df['Interest Rate']==0.03
```

Definan el nuevo DataFrame basándose en los filtros de año e interés que han creado.

```python
tres_porciento_30_a=df.loc[(tres_porciento) & (thty_year)]
tres_porciento_30_a
```

```python
five_pct_30_a=df.loc[(~tres_porciento) & (thty_year)]
five_pct_30_a
```

Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum(). ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?

```python
tres_porciento_30_a_cumsum=tres_porciento_30_a['Interest Paid'].cumsum()
tres_porciento_30_a_cumsum[:3]
```

Usen plt.plot(x, y, c=color) para trazar la línea negra (5 %). Queremos los valores x para que sean ['Month'] y los valores y para que sean la suma acumulativa del Interest Paid (interés pagado). Podemos usar el código de la pregunta anterior para definir los valores y dentro de la función plt.plot(). Pista: para el negro, c = 'k'

```python
cinco_porciento_30_a_cumsum=five_pct_30_a['Interest Paid'].cumsum()
```

```python
import matplotlib.pyplot as plt


# Extraer los valores x (meses) e y (suma acumulativa de interés pagado)
x = five_pct_30_a['Month']
y = cinco_porciento_30_a_cumsum

# Trazar la línea negra (5 %) utilizando plt.plot()
plt.plot(x, y, c='k', label='5%')
plt.plot()
x=five_pct_30_a['Month']
y=tres_porciento_30_a_cumsum
plt.plot(x, y, c='b', label='3%')
plt.plot()
# Añadir etiquetas a los ejes y un título al gráfico (opcional)
plt.xlabel('Month')
plt.ylabel('Cumulative Interest Paid')
plt.title('Cumulative Interest Paid over Time (5%)')

# Mostrar

plt.show()
```
