```python
def format_drive_url(url):
    file_id = url.split('/')[-2]
    download_url = f'https://drive.google.com/uc?id={file_id}&export=download'
    return download_url
```

```python
import pandas as pd
url=format_drive_url('https://docs.google.com/spreadsheets/d/1zdBJKtA9WdmBSqPkc5BdSLxohNsXce1q/edit?usp=sharing&ouid=106773039522351372029&rtpof=true&sd=true')
df=pd.read_excel(url)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Percentage of Dog Owners'], bins=10, alpha=0.5, color='blue', label='Dog Owners')
plt.xlabel('Percentage of Dog Owners')
plt.ylabel('Frequency')
plt.title('Histogram of Dog Ownership Percentage')
plt.show()

plt.hist(df['Percentage of Cat Owners'], bins=10, alpha=0.5, color='orange', label='Cat Owners')
plt.xlabel('Percentage of Cat Owners')
plt.ylabel('Frequency')
plt.title('Histogram of Cat Ownership Percentage')
plt.legend()
plt.show()

```

```python
# Trazar un histograma del porcentaje de dueños de perros
plt.hist(df['Percentage of Dog Owners'], bins=10, alpha=0.5, label='Dog Owners')

# Trazar un histograma del porcentaje de dueños de gatos
plt.hist(df['Percentage of Cat Owners'], bins=10, alpha=0.5, label='Cat Owners')

plt.xlabel('Percentage of Pet Owners')
plt.ylabel('Frequency')
plt.title('Percentage of Animal Owners')
plt.legend()
plt.show()
```

El rango de porcentajes de dueños de perros es más amplio que el de dueños de gatos. Esto significa que hay una mayor variabilidad en el porcentaje de dueños de perros entre los estados en comparación con el porcentaje de dueños de gatos.

En general, parece que el porcentaje de dueños de perros suele ser mayor que el porcentaje de dueños de gatos.

```python
data = [df['Mean Number of Cats'],df['Mean Number of Dogs per household']]
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot(data,
           notch = True,
           labels=['Cats','Dogs'],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          );
# The more you understand any library, the more you can do
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('Red')
plt.ylabel('Mean Number of Pets per State', fontsize = 20)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.title('Mean Number of  Household')
plt.show()
```

Los graficos indican que, en general, el promedio de gatos por hogares es mayor que el de perros.
