
Definir las características (X) y el objetivo (y)

```python
df_train_split=df.copy()
```

```python
X=df_train_split.drop(columns=['charges'])
y=df_train_split['charges']
```

Realizar un train test split a los datos para prepararlos para el aprendizaje automático


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Realizar una codificación one-hot cualquier característica nominal

```python
cat_selector = make_column_selector(dtype_include='object')
```

```python
train_cat_data = X_train[cat_selector(X_train)]
test_cat_data = X_test[cat_selector(X_test)]
train_cat_data
```

```python
# instancien la codificación one-hot
ohe_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
# ajusten el OneHotEncoder en los datos de entrenamiento
ohe_encoder.fit(train_cat_data)
# transforme los datos de entrenamiento y de prueba
train_ohe = ohe_encoder.transform(train_cat_data)
test_ohe = ohe_encoder.transform(test_cat_data)
train_ohe
```

```python
ohe_column_names = ohe_encoder.get_feature_names_out(train_cat_data.columns)
train_ohe = pd.DataFrame(train_ohe, columns=ohe_column_names)
test_ohe = pd.DataFrame(test_ohe, columns=ohe_column_names)
train_ohe
```

Escalar cualquier característica numérica

```python
X_train.describe().round(0)
```

```python
numeric_columns=X_train.select_dtypes(include=['int64', 'float64']).columns
```

```python
# instancien el escalador
scaler = StandardScaler()
# ajusten el escalador en los datos de entrenamiento
scaler.fit(X_train[numeric_columns])
```

```python
X_train_num=X_train[numeric_columns]
X_test_num=X_test[numeric_columns]
```

```python
train_scaled = scaler.transform(X_train_num)
# transformen los datos de prueba
test_scaled = scaler.transform(X_test_num)
# observen las 5 primeras filas de train_scaled
train_scaled[:5]
```

```python
X_train_scaled = pd.DataFrame(train_scaled, columns=X_train_num.columns)
X_train_scaled.head()
```

```python
X_test_scaled = pd.DataFrame(test_scaled, columns=X_test_num.columns)
```

Concatenar todas las características de vuelta a un DataFrame.
